# Tuple
They work exactly like lists, except that tuples can’t be changed inplace (they’re immutable) and are usually written as a series of items in parentheses,
not square brackets.

```python
t1=(2,3,4,7,9,10,12)
print(t1)
```

### Access elements using index

```python
print("First elements: ",t1[0])
print("last elements: ",t1[0])
print("First three elements: ",t1[0:3])
print("last three elements: ",t1[-3:])
```

# Dictionary
dictionaries are perhaps the most flexible built-in data type in Python.
* Accessed by key, not offset
* Unordered collections of arbitrary objects
* Variable-length, heterogeneous, and arbitrarily nestable
* Of the category “mutable mapping”
* Tables of Object Reference

### Empty Dictionary

```python
d1={}
d2=dict()
print(d1)
print(d2)
```

### Dictionary items accessed through keys

```python
d1={'player1':34,'player2':45,'player3':87}
print(d1)
```

```python
print(d1['player1'])
```

### Update dictionary

```python
d1['player1']=100
print(d1)
```

### Nested Dictionary

```python
score={'match1':{'player1':74,'player2':10,'player3':23},'match2':{'player1':30,'player3':82},'match3':{'player1':25,'player2':10,'player3':75}}
print(score)
```

```python
print(score['match3'])
```

### Access nested values

```python
print(score['match2']['player1'])
```

```python
d2=dict(play1=55,play2=71)
print(d2)
```

### keywords, zipped pairs, key lists

```python
player=['Rishab','Naraya','Debojyoti']
runs=[56,45,23]
d=dict(zip(player,runs))
print(d)
```

### String Method
strings provide a set of methods that implement more sophisticated text-processing tasks. Methods are simply functions that are associated
with particular objects.

```python
s='abc'
print(s.capitalize())
print(s)
```

```python
s='abc'
print(s.upper())
print(s)
```

```python
s='MCA25'
for item in s:
  print(f'{item} is alphabet:',item.isalpha())
  print(f'{item} is digit:',item.isdigit())
```
