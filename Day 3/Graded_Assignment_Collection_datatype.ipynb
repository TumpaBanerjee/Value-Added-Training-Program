Practice Program:
Write a program to apply a 1d filter of filter from left to right by
multiplying the corresponding values and then adding on a signal.

```python
def filteringSignal(sig,fltr):
    """
This function will apply the filter fltr from left to right by
multiplying the corresponding values and then adding
"""
    lengthS=len(sig)
    lengthF=len(fltr)
    filteredS=list()
    for i in range(lengthS-lengthF+1):
        fltrd=0
        for ind in range(lengthF):
            fltrd+=sig[i+ind]*fltr[ind]
        filteredS.append(fltrd)
        print(filteredS)
```

```python
filteringSignal([1,2,3,4,5,6,7,8],[-1,1,-1])
```

```python
filteringSignal([1,2,13,4,5,11,7,8],[-1,0,1])
```
