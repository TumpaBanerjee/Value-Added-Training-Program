# Core Conditional Statements
---
The primary conditional control statements are:

$if$ statement: Executes a block of code only if its condition is $True$

```python
num=20
if num%2==0:
  print("The given number is even")
print("This statement is outside of of if block")
```

20%2=0, control execute if block

```python
num=13
if num%2==0:
  print("The given number is even")
print("This statement is outside of of if block")
```

20%2=1!=0, control does not execute if block

## if-else statement: Executes one block if the condition is $True$ and a different block if the condition is $False$.

```python
num=13
if num%2==0:
  print("The given number is even")
else:
  print("Odd Number")
print("This statement is outside of of if block")
```

13%2=1!=0, control does not execute if block, control goes to else block and print "Odd Number"

## if-elif-else statement: Allows checking multiple conditions sequentially. Execution stops at the first condition that evaluates to $True$.

```python
num=0
if num >0:
  print("Positive number")
elif num<0:
  print("Negative number")
else:
  print("Zero")
```

## Nested if
if within if

```python
acc_balance=-20000
if acc_balance<0:
  if -acc_balance >= 1000:
    print("You are requested to pay your loan amount with 10% interest")
  else:
    print("Your are requested to pay your due amount")
else:
  print("You do not have any due in this month")
```
